```python
dbutils.widgets.text("run_date", "")
run_date = dbutils.widgets.get("run_date")

GOLD_NOTEBOOKS = [
  "/Workspace/Users/ghanemkhal@gmail.com/Databricks-Lakehouse-Project/bike_lakehouse_2026/Gold/gold_customers",
  "/Workspace/Users/ghanemkhal@gmail.com/Databricks-Lakehouse-Project/bike_lakehouse_2026/Gold/gold_products",
  "/Workspace/Users/ghanemkhal@gmail.com/Databricks-Lakehouse-Project/bike_lakehouse_2026/Gold/gold_sales",
]

TIMEOUT_SECONDS = 0

results = []

for nb in GOLD_NOTEBOOKS:
  print(f"\n=== Running Gold notebook: {nb} ===")
  try:
    out = dbutils.notebook.run(
      nb,
      timeout_seconds=TIMEOUT_SECONDS,
      arguments={"run_date": run_date}
    )
    results.append((nb, "SUCCESS", out))
    print(f" SUCCESS: {nb} | output: {out}")
  except Exception as e:
    results.append((nb, "FAILED", str(e)))
    print(f" FAILED: {nb}\n{e}")
    raise

summary = "\n".join([f"{s} - {p}" for p, s, _ in results])
dbutils.notebook.exit("Gold orchestration completed.\n" + summary)

    
```
